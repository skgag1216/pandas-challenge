```python
import pandas as pd
```

```python
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
#District Summary
school_data_complete.head(-1)
```

```python
all_schools = school_data_complete["school_name"].value_counts()
total_schools = all_schools.count()
total_schools
```

```python
total_students = school_data_complete["student_name"].count()
total_students
```

```python
budget_df = school_data_complete.filter(['school_name', 'budget'])
#budget_df.head(-1) to view data
```

```python
working_budget_df = budget_df.drop_duplicates(subset = "school_name", keep = 'first', inplace = True)
```

```python
total_budget = budget_df["budget"].sum()
total_budget.astype("float")
total_budget
```

```python
#print(type(total_budget)) #data type is int here, change to float
```

```python
avg_math = round((school_data_complete["math_score"].mean()), 2)
avg_math
```

```python
avg_reading = round((school_data_complete["reading_score"].mean()), 2)
avg_reading
```

```python
num_pass_math = school_data_complete.loc[(school_data_complete["math_score"] >= 70)]
percent_pass_math = round((num_pass_math["math_score"].count() / total_students * 100), 2)
percent_pass_math
```

```python
num_pass_reading = school_data_complete.loc[(school_data_complete["reading_score"] >= 70)]
percent_pass_reading = round((num_pass_reading["reading_score"].count() / total_students * 100), 2)
percent_pass_reading
```

```python
num_pass_mr = school_data_complete.loc[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)]
percent_pass_mr = round((num_pass_mr["student_name"].count() / total_students * 100), 2)
percent_pass_mr
```

```python
district_data = {'Total Schools': [total_schools], 'Total Students': [total_students],'Total Budget': [total_budget], 
                'Average Math Score': [avg_math], 'Average Reading Score': [avg_reading], '% Passing Math': [percent_pass_math], 
                '% Passing Reading': [percent_pass_reading], '% Overall Passing': [percent_pass_mr]}
```

```python
district_df = pd.DataFrame(district_data)
district_df["Total Budget"] = district_df["Total Budget"].astype("float").map('${:,.2f}'.format)
district_df["Total Students"] = district_df["Total Students"].astype("int").map('{:,}'.format)
```

```python
reduced_school_df = school_data_complete.loc[:, ["school_name", "student_name", "reading_score", "math_score", "type", "size", "budget"]] 
reduced_school_df
```

```python
school_index_df = school_data.set_index("school_name")
school_index_df = school_index_df.drop(["School ID"], axis=1)
school_index_df.sort_index() 
# gather new variables for type, size, budget, number of students and budget per student from indexed school df
typeof_school = school_index_df["type"]
number_students_per_school = school_index_df["size"] #returns number of students per school
budgetof_school = school_index_df["budget"]
budget_per_student = budgetof_school / number_students_per_school
budget_per_student_clone = budgetof_school / number_students_per_school #for formatting purposes, this one has NO formatting
# budget_per_student #(run this calc alone to make sure numbers are correct)

# pull math and reading scores per school by using groupby() from the merged school/student data 
# make sure new df of complete data uses school as index just like above
reduced_schoolas_index_df = reduced_school_df.set_index("school_name")
grouped_by_school_df = reduced_schoolas_index_df.groupby("school_name")

school_avg_math = grouped_by_school_df["math_score"].mean() #outputs the average math score per school
school_avg_reading = grouped_by_school_df["reading_score"].mean() #outputs the average reading score per school
#can run each of these individually to make sure output is correct

#calc percent of students that passed math per school
students_per_school = grouped_by_school_df["student_name"].count() #pull total number of students per school, 
                                                                #can run this and compare to school_data df to verify output
grouped_pass_math = num_pass_math.groupby("school_name") #refer to district df calculations to pull this num_pass variable
perc_passed_math_per_school = grouped_pass_math["student_name"].count() / students_per_school * 100
#perc_passed_math_per_school (run individually to verify output)

#percent of students that passed reading per school
grouped_pass_reading = num_pass_reading.groupby("school_name") #refer to district df calculations to pull this num_pass variable
perc_passed_read_per_school = grouped_pass_reading["student_name"].count() / students_per_school * 100
#perc_passed_read_per_school

#calc percent of students that passed math and reading per school
grouped_pass_mr = num_pass_mr.groupby("school_name") #refer to district df calculations to pull this num_pass variable
perc_passed_mr_per_school = grouped_pass_mr["student_name"].count() / students_per_school * 100
#perc_passed_mr_per_school

school_summary_df = pd.DataFrame({"School Type" : typeof_school,
                                 "Total Students" : number_students_per_school,
                                 "Total School Budget": budgetof_school,
                                 "Per Student Budget Clone no format": budget_per_student_clone,
                                 "Per Student Budget": budget_per_student,
                                 "Average Math Score": school_avg_math,
                                 "Average Reading Score": school_avg_reading,
                                 "% Passing Math": perc_passed_math_per_school,
                                 "% Passing Reading": perc_passed_read_per_school,
                                 "% Overall Passing": perc_passed_mr_per_school})

school_summary_df["Total School Budget"] = school_summary_df["Total School Budget"].astype("float").map('${:,.2f}'.format)
school_summary_df["Per Student Budget"] = school_summary_df["Per Student Budget"].astype("float").map('${:,.2f}'.format)
school_summary_df["Total Students"] = school_summary_df["Total Students"].astype("int")
school_summary_df[["School Type",
                  "Total Students",
                  "Total School Budget",
                  "Per Student Budget",
                  "Average Math Score",
                  "Average Reading Score",
                  "% Passing Math",
                  "% Passing Reading",
                  "% Overall Passing"]]
```

```python
# Sort and display the top five performing schools by % overall passing
top_performing_schools = school_summary_df.sort_values(by=["% Overall Passing"], ascending=False)
top_performing_schools.head()
```

```python
lowest_performing_schools = school_summary_df.sort_values(by=["% Overall Passing"], ascending=True)
lowest_performing_schools.head()
```

```python
student_data # start with student_data 
student_school_index = student_data.set_index("school_name")
ninth_grade = student_school_index.loc[student_school_index["grade"] == "9th"].groupby(["school_name"])["reading_score"].mean()
tenth_grade = student_school_index.loc[student_school_index["grade"] == "10th"].groupby(["school_name"])["reading_score"].mean()
eleventh_grade = student_school_index.loc[student_school_index["grade"] == "11th"].groupby(["school_name"])["reading_score"].mean()
twelfth_grade = student_school_index.loc[student_school_index["grade"] == "12th"].groupby(["school_name"])["reading_score"].mean()

student_reading_scores_bygrade = pd.DataFrame({"9th": ninth_grade,
                          "10th": tenth_grade,
                          "11th": eleventh_grade,
                          "12th": twelfth_grade})
student_reading_scores_bygrade
```

```python
bins = [0, 584, 630, 645, 680]
spending_ranges = ["<$585", "$585-$630", "$630-$645", "$645-$680"]
spending_range_summary = school_summary_df.copy()
spending_range_summary = spending_range_summary.reset_index()

spending_range_summary["Spending Ranges (Per Student) "] = pd.cut(spending_range_summary["Per Student Budget Clone no format"], bins, labels = spending_ranges)
spending_range_summary = spending_range_summary.groupby(["Spending Ranges (Per Student) "]).mean()
spending_range_summary = spending_range_summary[["Average Math Score", "Average Reading Score", 
                                                 "% Passing Math", "% Passing Reading", 
                                                "% Overall Passing"]]
```

```python
spending_range_summary
```

```python
size_bins = [0, 999, 1999, 5000]
size_labels = ["Small(<1000)", "Medium(1000-2000)", "Large(2000-5000)"]
spending_size_summary = school_summary_df.copy()
spending_size_summary = spending_size_summary.reset_index()
spending_size_summary["School Size"] = pd.cut(spending_size_summary["Total Students"], size_bins, labels = size_labels)
spending_size_summary = spending_size_summary.groupby("School Size").mean()
spending_size_summary = spending_size_summary[["Average Math Score", "Average Reading Score", 
                                                 "% Passing Math", "% Passing Reading", 
                                                "% Overall Passing"]]
spending_size_summary
```
